```python
galah_dr3 = Table.read('../../catalogs/GALAH_DR3_main_allspec_v1.fits')
```

```python
selection = (
    galah_dr3['flag_sp'] <= 256
    )
```

```python
def vmic_DF16(teff, logg):
    """
    https://ui.adsabs.harvard.edu/abs/2016A%26A...585A..75D/abstract
    """
    return 0.998+3.16*10**(-4)*(teff-5500.)-0.253*(logg-4.0)-2.86*10**(-4)*(teff-5500)*(logg-4.0)+0.165*(logg-4.0)**2.
```

```python
def vmic_Lind(teff, logg):
    
    # first for non-giants:
    ai1 = 1.1
    ai2 = 1.0*10**(-4)
    ai3 = 4*10**(-7)

    vmic = ai1 + ai2*(teff-5500.) + ai3 * (teff-5500.)^2    
    
    ai1 = 1.1
    ai2 = 1.6*10**(-4)
    ai3 = 0
    
    vmic_cool_dwarfs = ai1 + ai2*(teff-5500.) + ai3 * (teff-5500.)^2    
    
    cool_dwarfs = (logg > 4.2) & (teff < 5500)
    
    vmic[cool_dwarfs] = vmic_cool_dwarfs[vmic_cool_dwarfs]
    
    return vmic
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(15,5))#,sharex=True, sharey=True)

kwargs = dict(
    bins=100,
    norm=LogNorm(),
    cmin=1,
    cmap=parula
)
p,x,y,s = ax1.hist2d(
    galah_dr3['teff'][selection],
    galah_dr3['vmic'][selection],
    range=[[3000,8500], [0, 4]],
    **kwargs
    )
c = plt.colorbar(s, ax=ax1)
c.set_label('Nr. Spectra')

p,x,y,s = ax2.hist2d(
    galah_dr3['teff'][selection],
    vmic_DF16(
        galah_dr3['teff'][selection],
        galah_dr3['logg'][selection]
        ),
    range=[[3000,8500], [0, 4]],
    **kwargs
    )
c = plt.colorbar(s, ax=ax2)
c.set_label('Nr. Spectra')

p,x,y,s = ax3.hist2d(
    galah_dr3['teff'][selection],
    vmic_DF16(
        galah_dr3['teff'][selection],
        galah_dr3['logg'][selection]
        ) - galah_dr3['vmic'][selection],
    range=[[3000,8500], [-2.5,2.5]],
    **kwargs
    )
c = plt.colorbar(s, ax=ax3)
c.set_label('Nr. Spectra')

ax1.set_xlim(8500,3000)
ax2.set_xlim(8500,3000)
ax3.set_xlim(8500,3000)

ax1.set_xlabel('Teff GALAH DR3')
ax1.set_ylabel('vmic GALAH DR3')
ax2.set_xlabel('Teff GALAH DR3')
ax2.set_ylabel('vmic DF16')
ax3.set_xlabel('Teff GALAH DR3')
ax3.set_ylabel('vmic DF16-DR3')

plt.tight_layout()
plt.savefig('../../dr3_release_paper/figures/Vmic_comparisons.png',dpi=300,bbox_inches='tight')
```
